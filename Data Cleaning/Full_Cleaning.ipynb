```python
import pandas as pd
from pathlib import Path
pd.set_option('display.max_columns',None)
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option("display.max_rows", None)
from ast import literal_eval
```

```python
filepath = Path("Downloads/Cleaned_Main_Data.csv")
filepath2 = Path("Downloads/cleaned_mal_studio_data.csv")
main = pd.read_csv(filepath)
studio = pd.read_csv(filepath2)


```

```python
main = main.rename(columns = {"title":"Name"})
```

```python
df = pd.merge(main,studio, how = "left", on = ["Name", "Name"])
```

```python
df.head()
```

```python
df.columns
```

```python
cols = ['uid', 'Name', 'genre', 'aired', 'episodes', 'members',
       'popularity', 'ranked', 'score','Type',
       'Studio', 'Genres']
df = df.loc[:,cols]
```

```python
df.head()
```

```python
df.info()
```

```python
df.head()
```

```python
air_date_split = df['aired'].str.split(' to ')
df['Start Date'] = air_date_split.apply(lambda x : x[0])
df['End Date'] = air_date_split.apply(lambda x : x[1] if len(x) > 1 else None)
```

```python
df.drop('aired', axis = 1, inplace = True)
```

```python
df.head()
```

```python
df.head()
```

```python
df.info()
```

```python
df2 = df.drop(df[df["Start Date"] == "Not available"].index)
```

```python


df2['Start Date'] = pd.to_datetime(df2['Start Date'])
# #add year of release column
```

```python
cols = ['uid', 'Name', 'genre', 'episodes', 'members',
       'popularity', 'ranked', 'score','Type',
       'Studio','Start Date','End Date']
df2 = df.loc[:,cols]
```

```python
df2.head()
```

```python
df2.info()
```

```python
df2["Studio"] = df2.Studio.apply(lambda x:literal_eval(x)[0] if pd.notnull(x) else x)
```

```python
df2["genre"] = df2.genre.apply(lambda x:literal_eval(x) if pd.notnull(x) else x)
df2["genre"] = df2.genre.apply(lambda x:x[0] if len(x) >0  else None)
```

```python
df2.head()
```

```python
df2.to_csv('Cleaned_Data_With_Year.csv')
```
